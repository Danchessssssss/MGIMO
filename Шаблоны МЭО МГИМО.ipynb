# Тип 2

## Для датасета TableF5-1 оцените probit-регрессию LFP~WA+I(WA**2)+CIT+UN+np.log(FAMINC). Тестируется значимость коэффициента при WA. 
## Укажите тестовую статистику, критическое значение и сделайте вывод. Ответы округлите до 2 десятичных знаков. Уровень значимости 1%

## Импорт библиотек:

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy import stats
```

## Загрузка и подготовка данных:
Замените 'path_to_your_data.csv' на путь к вашему файлу данных.

```python
df = pd.read_csv('TableF5-1.csv')
df['log_FAMINC'] = np.log(df['FAMINC'])
df
```

## Оценка probit-регрессии:

```python
model = sm.Probit.from_formula('LFP ~ WA + I(WA**2) + CIT + UN + log_FAMINC', data=df)
result = model.fit()
```

## Проведение теста на значимость коэффициента при WA:

```python
# Получение стандартной ошибки коэффициента WA
se = result.bse['WA']
# Получение коэффициента при WA
coef = result.params['WA']
# Расчет z-статистики
z_stat = coef / se
# Расчет p-значения
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
# Вывод результатов
print(f"Тестовая статистика: {z_stat:.2f}")
print(f"P-значение: {p_value:.2f}")
```

## Определение критического значения и вывод:

```python
critical_value = stats.norm.ppf(1 - 0.01/2)
print(f"Критическое значение: {critical_value:.2f}")
```

Для вывода: 

Если |тестовая статистика| > критическое значение, коэффициент при WA статистически ### значим.

В противном случае, коэффициент не является статистически значимым.


# Тип 3

## Для датасета TableF5-1 оцените logit-регрессию LFP~K618+CIT+UN и тестируйте её значимость. Укажите тестовую статистику, критическое значение и сделайте вывод. Ответы округлите до 2 десятичных знаков. Уровень значимости 10%

## Импорт библиотек:

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy.stats import chi2
```

## Загрузка и подготовка данных:

```python
df = pd.read_csv('TableF5-1.csv')
df
```

## Оценка logit-регрессии:

```python
model = sm.Logit(df['LFP'], df[['K618', 'CIT', 'UN']])
result = model.fit()
```

## Проведение теста на общую значимость модели (Likelihood Ratio Test):
Сначала оцениваем модель только с константой:

```python
model_const = sm.Logit(df['LFP'], np.ones(len(df)))
result_const = model_const.fit()
```

Теперь проводим тест:

```python
lr_stat = 2 * (result.llf - result_const.llf)
p_value = chi2.sf(lr_stat, df=result.df_model - result_const.df_model)
print(f"Тестовая статистика (LR): {lr_stat:.2f}")
print(f"P-значение: {p_value:.2f}")
```

## Определение критического значения:
Для уровня значимости 10% и числа степеней свободы, равного разнице в количестве параметров между полной и урезанной моделью:

```python
df_freedom = result.df_model - result_const.df_model
critical_value = chi2.ppf(0.9, df_freedom)
print(f"Критическое значение: {critical_value:.2f}")
```

## Вывод о значимости регрессии:

```python
if lr_stat > critical_value:
    print("Регрессия значима.")
else:
    print("Регрессия не значима.")
```
